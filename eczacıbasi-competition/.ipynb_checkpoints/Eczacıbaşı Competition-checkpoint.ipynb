```python
import pandas as pd
import numpy as np
from datetime import datetime
from pandas_profiling import ProfileReport
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, TweedieRegressor, SGDRegressor
from sklearn.svm import SVR, LinearSVR
from sklearn.neighbors import KNeighborsRegressor
from catboost import CatBoostRegressor
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.ensemble import ExtraTreesRegressor, IsolationForest
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import xgboost as xgb
```

# Reading the data

```python
data = pd.read_excel("Takmlarla Paylalacak Veri Seti.xlsx")
data_to_predict = pd.read_excel("id.xlsx")
```

```python
data = data.groupby(['date',"customer","item"]).agg({'order': sum}).reset_index()
```

# Kmeans clustering of the items

```python
data_for_kmeans = data.groupby(["item"]).agg({'order': sum}).reset_index()
data_for_kmeans = pd.concat([data_for_kmeans, pd.get_dummies(data_for_kmeans["item"], drop_first=True)],1).drop("item",1)
data_for_kmeans["order"] = np.log(data_for_kmeans["order"])
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)
X = data_for_kmeans

for k in K:
	# Building and fitting the model
	kmeanModel = KMeans(n_clusters=k).fit(X)
	kmeanModel.fit(X)

	distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_,
										'euclidean'), axis=1)) / X.shape[0])
	inertias.append(kmeanModel.inertia_)

	mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_,
								'euclidean'), axis=1)) / X.shape[0]
	mapping2[k] = kmeanModel.inertia_

plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4).fit(data_for_kmeans)
data_for_kmeans["labels"] = kmeans.labels_
```

```python
kmeans_results = data.groupby(["customer", "item"]).agg({'order': sum}).reset_index()
kmeans_results["labels"] = kmeans.labels_
kmeans_results.drop("order",1,inplace=True)
kmeans_results.drop("customer",1,inplace=True)
data =pd.merge(data, kmeans_results, on="item")
```

# Adding time features

```python
data["date"] = data["date"].apply(lambda x: datetime.strptime(x, "%Y-%m"))


data_final = data.sort_values(by=["date","customer","item"]).reset_index(drop=True)


#data_final = pd.concat([data, pd.get_dummies(data["customer"], drop_first=True)],1)


#data_final = pd.concat([data_final, pd.get_dummies(data["item"], drop_first=True)],1).drop("item",1)


data_final["year"] = data_final["date"].apply(lambda x: x.year)


data_final["month"] = data_final["date"].apply(lambda x: x.month)


data_final["sin_month"] = data_final["month"].apply(lambda x: np.sin(x*np.pi/6))


data_final["cos_month"] = data_final["month"].apply(lambda x: np.cos(x*np.pi/6))


data_final["tan_month"] = data_final["month"].apply(lambda x: np.tan(x*np.pi/6))


#data_final["new_year"] = data_final["year"].apply(lambda x: x-2018)


#data_final = pd.concat([data_final, pd.get_dummies(data_final["labels"], prefix="label_", drop_first=True)],1).drop("labels",1)


data_final["season"] = data_final["month"]%12 // 3 + 1


#data_final = pd.concat([data_final, pd.get_dummies(data_final["season"], prefix="season_", drop_first=True)],1).drop("season",1)


data_final["is_covid"] = data_final["date"].apply(lambda x: 0 if x<datetime(2020,3,1) else (1 if x<datetime(2020,5,1) else 2))


#data_final.drop(["year","month", "customer"], 1, inplace=True)
```

# Adding Turkish export data

```python
ihracat = pd.read_excel("tsf serkap.xlsx")
ihracat["date"] = ihracat["Yıl"].astype(str) + "-" + ihracat["Ay"].astype(str)
ihracat["date"] = ihracat["date"].apply(lambda x: datetime.strptime(x, "%Y-%m"))
ihracat.drop(["Ay","Yıl"],1,inplace=True)
ihracat = ihracat[['date','İhracat Kg', 'İhracat m2', 'İthalat Kg', 'İthalat m2', 'İhracat Dolar','İhracat Euro', 'İthalat Dolar', 'İthalat Euro']]
ihracat.sort_values(by="date", inplace=True)
ihracat.reset_index(drop=True, inplace=True)
data_final = pd.merge(data_final, ihracat, on="date") 
```

# Train-test separation 

```python

data_train = data_final[data_final["date"] < '2020-05-01'][['date', 'customer', 'item', 'order', 'labels', 'year', 'sin_month',
       'cos_month', 'season', 'is_covid']]
data_test = data_final[data_final["date"] >= '2020-05-01'][['date', 'customer', 'item', 'order', 'labels', 'year', 'sin_month',
       'cos_month', 'season', 'is_covid']]

data_train.drop("date", 1, inplace=True)
data_test.drop("date", 1, inplace=True)

y_train = data_train["order"]
X_train = data_train.drop("order",1)

y_test = data_test["order"]
X_test = data_test.drop("order",1)


```

# CatBoost model (dummy variables are not necessary, model takes a parameter of categorical features)

```python
model = CatBoostRegressor(cat_features = ['customer', 'item', 'year','season','labels','is_covid'],
                          iterations=3000, learning_rate=0.01)
model.fit(X_train, y_train, eval_set=(X_test, y_test), plot=True, sample_weight=s_w)
#np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
pred = model.predict(X_test)
#pred = np.exp(pred)
print(np.sqrt(mean_squared_error(y_test, pred)))
```

```python
data_final = pd.concat([data_final, pd.get_dummies(data_final["customer"],
                                                   prefix="customer", drop_first=True)], axis=1).drop("customer", axis=1)
data_final = pd.concat([data_final, pd.get_dummies(data_final["item"],
                                                   prefix="item", drop_first=True)], axis=1).drop("item", axis=1)
data_final = pd.concat([data_final, pd.get_dummies(data_final["season"],
                                                   prefix="season", drop_first=True)], axis=1).drop("season", axis=1)
data_final = pd.concat([data_final, pd.get_dummies(data_final["labels"],
                                                   prefix="label", drop_first=True)], axis=1).drop("labels", axis=1)
```

```python
data_final.drop(columns_to_drop, axis=1, inplace=True)
```

```python

data_train = data_final[data_final["date"] < '2020-05-01']
data_test = data_final[data_final["date"] >= '2020-05-01']

data_train.drop("date", 1, inplace=True)
data_test.drop("date", 1, inplace=True)

y_train = data_train["order"]
X_train = data_train.drop("order",1)

y_test = data_test["order"]
X_test = data_test.drop("order",1)


```

```python
model = LinearRegression()
model.fit(X_train, y_train)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = Ridge()
model.fit(X_train, y_train)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = Lasso(positive=True)
model.fit(X_train, y_train, sample_weight=s_w)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = TweedieRegressor(power=1, alpha=0.2, link="log", max_iter=10000)
model.fit(X_train, y_train, sample_weight=s_w)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = SGDRegressor(shuffle=False, learning_rate='adaptive', max_iter=20000)
model.fit(X_train, y_train, sample_weight=s_w)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = KNeighborsRegressor(n_neighbors=3)
model.fit(X_train, y_train)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = LinearSVR(C=1, loss= "squared_epsilon_insensitive", intercept_scaling=0.1)
model.fit(X_train, y_train)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
model = SVR()
model.fit(X_train, y_train)
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```
